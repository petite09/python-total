##  Descripción Proyecto 2:

Situación: Tú trabajas en una empresa donde los vendedores reciben comisiones del 13% por sus ventas totales. Tu jefe quiere que ayudes a los vendedores a calcular sus comisiones creando un pregunte su nombre y cuánto han vendido en este mes.

Tu programa le va a responder con una frase que incluya su nombre y el monto que le corresponde por las comisiones.

“Ok Juan. Este mes ganaste $650”

Tips:

- Programa debe comenzar preguntando cosas al usuario: ¿Nombre? ¿Ventas? → se necesita input
- Y deberías usar variables para almacenar esos ingresos.
- *Recordar que los ingresos de usuario se almacenan como str, por lo tanto, se debe convertir uno de esos ingresos en un float para poder hacer operaciones.*
- Operaciones: Calcular el 13% del número que haya ingresado el usuario (*13/100) (almacenar resultado en una variable)
- Al imprimir el resultado en pantalla asegurarsede que esa información no tenga más de 2 decimales → round(13% de la variable,2) para que sea fácil de leer y luego organiza todo eso en un str al que debes dar formato.

## Proyecto día 2

```python
nombre = input("Nombre: ")
ventas = input("Ventas: ")

#Cambiar tipo de dato de ventas (str) a float
ventas = float(ventas)
ganancia = round(ventas*13/100,2)

print(f"Ok {nombre}. Este mes ganaste ${ganancia}")
```
