###  Proyecto 1:

Situación: Tu mejor amigo ha puesto una fábrica de cerveza y tiene todo listo, pero le falta la identidad. No se le ocurre qué nombre ponerle a su cerveza para que tenga una identidad única y original.

Crear un programa que te va a hacer 2 preguntas y luego te va a decir cuál es el nombre de tu cerveza.

Crear un código en Python que le pida a tu amigo que responda dos pregunta que requieran una sola palabra cada una y que luego le muestre en pantalla esas palabras combinadas, para formar una marca creativa. Puedes usar las preguntas que quieras.

**Desafío extra**: Que el nombre de la cerveza se imprima entre comillas y que la impresión final quede en 2 líneas.





```python
# Desarrollo

print("La cerveza se llama: " + input("¿Cuál es tu tipo de cerveza favorito? ") + " " + input("Nombra tu color favorito en inglés: "))
```

Falta lograr poner las respuestas en una línea aparte y las comillas en la respuesta.

```python
print("La cerveza se llama:\n"+ input("¿Cuál es tu tipo de cerveza favorito? ") + " " + input("Nombra tu color favorito en inglés: "))
```

Ahora solo falta poner las comillas en la respuesta.



### El programa final queda:

```python
print("La cerveza se llama:\n"+ "'" + (input("¿Cuál es tu tipo de cerveza favorito? ")) + " " + (input("Nombra tu color favorito en inglés: ")) +"'")
```
